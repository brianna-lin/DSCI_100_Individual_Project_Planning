#### Summary Statistics

```r
summary(players_data)
```

```r
player_genders <- players_data |>
    group_by(gender) |>
    summarize(count = n())
player_experience <- players_data |>
    group_by(experience) |>
    summarize(count = n())
player_subscribers <- players_data |>
    group_by(subscribe) |>
    summarize(count = n())
player_genders
player_experience
player_subscribers
```

By observing the statistics above, we can see that for this data set, there are more subscribers than non-subscribers, most players are amateur level, and most are male. We also know that the max age of the players is 99 while the mean is 21.28 years old. The max number of hours played is 223.10 while the mean is 5.85 hours.

Another issue that can be identified from the statistics above is that if we were to use the data for a classification problem, some of the variables in this dataset exhibit oversampling (e.g. more data on male players than other gender players), which would lower the accuracy of a KNN classification model.

```r
experience_summary <- players_data |>
    group_by(experience) |>
    summarise(avg_played_hrs = mean(played_hours, na.rm = TRUE), max_hrs_played = max(played_hours, na.rm = TRUE),
              avg_age = mean(age, na.rm = TRUE), max_age = max(age, na.rm = TRUE))

subscriber_summary <- players_data |>
    group_by(subscribe) |>
    summarise(avg_played_hrs = mean(played_hours, na.rm = TRUE), max_hrs_played = max(played_hours, na.rm = TRUE),
              avg_age = mean(age, na.rm = TRUE), max_age = max(age, na.rm = TRUE))

gender_summary <- players_data |>
    group_by(gender) |>
    summarise(avg_played_hrs = mean(played_hours, na.rm = TRUE), max_hrs_played = max(played_hours, na.rm = TRUE),
              avg_age = mean(age, na.rm = TRUE), max_age = max(age, na.rm = TRUE))

experience_summary
subscriber_summary
gender_summary
```
