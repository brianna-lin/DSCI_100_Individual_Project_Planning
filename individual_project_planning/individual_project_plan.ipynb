```r
library(repr)
library(tidyverse)
```

```r
players_data <- read_csv("data/players.csv")
sessions_data <- read_csv("data/sessions.csv")
```

```r
players_data
sessions_data
```

By looking at the data sets above of the players and sessions, we can see that for the players data set, there are 196 observations and 7 variables (experience level, whether the player is a subscriber, their hashed email, number of hours played, the player’s name, their gender, and their age). The sessions data set has 1535 observations and 5 variables (hashed email, start time of the session, end time, and the original start and end times).

The sessions data set is untidy as the columns labelled start_time and end_time contain two values in each cell (the date and the clock time). If we choose to answer our analysis question using this data set, we will need to tidy it first by separating the values in those two columns.

#### Summary Statistics

```r
summary(players_data)
```

```r
summary(sessions_data)
```

```r
player_genders <- players_data |>
    group_by(gender) |>
    summarize(count = n())
player_experience <- players_data |>
    group_by(experience) |>
    summarize(count = n())
player_subscribers <- players_data |>
    group_by(subscribe) |>
    summarize(count = n())
player_genders
player_experience
player_subscribers
```

```r
# separate the values in the start and end_time columns
sessions_tidy <- sessions_data |>
    separate(col = start_time,
             into = c("start_date", "start_time"),
             sep = " ") |>
    separate(col = end_time,
             into = c("end_date", "end_time"),
             sep = " ")
session_length_data <- sessions_tidy |>
    # make new columns displaying start and end time in minutes
    mutate(start_time_min = as.numeric(substr(start_time, 1, 2)) * 60 + as.numeric(substr(start_time, 4, 5)),
          end_time_min = as.numeric(substr(end_time, 1, 2)) * 60 + as.numeric(substr(end_time, 4, 5))) |>
    # subtract end minutes from start minutes to obtain length of sessions
    mutate(session_length = abs(end_time_min - start_time_min)) |>
    select(start_time, end_time, session_length)
#head(session_length_data)
options(repr.plot.width = 8, repr.plot.height = 8)
session_length_hist <- session_length_data |>
    ggplot(aes(x = session_length)) +
        geom_histogram(binwidth=60) +
        labs(x = "Session Length (min)", title = "Distribution of Length of Play Sessions") +
        theme(text = element_text(size = 20))
session_length_hist
```
